# load dataset


#### The dataset includes descriptions, price, sales-rank, brand info, and co-purchasing link about 9.4 million products.

```python
import pandas as pd
import numpy as np
import gzip

def parse(path):
  g = gzip.open(path, 'rb')
  for l in g:
    yield eval(l)

def getDF(path):
  i = 0
  df = {}
  for d in parse(path):
    df[i] = d
    i += 1
  return pd.DataFrame.from_dict(df, orient='index')

df = getDF('input/metadata.json.gz')
```

# Clean data

### Keep title and categories

```python
# the dataset is huge. Select useful data from it.
del df['asin']
del df['salesRank']
del df['imUrl']
del df['description']
del df['price']
del df['related']
del df['brand']
df.head(3)
```

```python
df["subcategory1"] = df['primary_category_list'].str[0]
```

```python
category = df.groupby('subcategory1').size()
category = category.rename('Number')
category = category.sort_values(ascending=False)
```

## drop 'books' category, because there seems no book in target data

```python
df = df[df.subcategory1 != 'Books']
df = df[df.subcategory1 != '']
```

```python
df = df[df.subcategory1 != '']
```

## we could tell that in this dataset, there exist a lot of music categories that seems not to exist in our target dataset. Also, the above data tells that these categories only take a small percent of purchase. Therefore, for simplicity, these small categories are ignored in my model.

```python
categoryList = category[category > 3000].index.tolist()
categoryList
```

```python
categoryList.remove('Pop')
categoryList.remove('Dance & Electronic')
categoryList.remove('Digital Music')
```

```python
df = df[df['subcategory1'].isin(categoryList)]
```

## seperate 'Clothing, Shoes & Jewelry' into three subcategories

```python
df[df['subcategory1'] == 'Clothing, Shoes & Jewelry']
```

```python
df["subcategory2"] = df['primary_category_list'].str[1]
df["subcategory3"] = df['primary_category_list'].str[2]
df["subcategory4"] = df['primary_category_list'].str[3]
```

```python
category = df[df['subcategory1'] == 'Clothing, Shoes & Jewelry'].groupby('subcategory3').size()
category = category.rename('Number')
category = category.sort_values(ascending=False)
category
```

```python
clothingShoesJewelry = ['Clothing','Accessories', 'Shoes', 'Jewelry', 'Watches', 'Costumes & Accessories']
```

```python
# define macro categories
df['macro categories'] = np.where(df['subcategory1'] == 'Clothing, Shoes & Jewelry', df["subcategory3"], df["subcategory1"])
```

```python
df.head()
```

```python
df = df[(df['subcategory1'] != 'Clothing, Shoes & Jewelry') | (df['subcategory3'].isin(clothingShoesJewelry))]
```

```python
df = df[df['macro categories'] != '']
```

```python
category = df.groupby('macro categories').size()
category = category.rename('Number')
category = category.sort_values(ascending=False)
category
```
